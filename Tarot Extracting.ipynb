```python
# Demo for exctracting textual associations of Tarot cards for NLP purposes
# Scrapes content from http://www.learntarot.com/
```

```python
# Importing all stuff
import requests
import html2text
import pprint
import re
import pandas as pd
from IPython.core.display import HTML, display
from string import Template
import json
```

```python
minorOrder = ['a','2','3','4','5','6','7','8','9','10','pg','kn','qn','kg']
# Generates urls for content extraction
# First the Major Arcana
urls =  ['http://www.learntarot.com/maj0' + str(i) + '.htm' for i in range(0,10)]
urls += ['http://www.learntarot.com/maj' + str(i) + '.htm' for i in range(10,22)]
# Wands
urls += ['http://www.learntarot.com/w' + card + '.htm' for card in minorOrder]
# Cups
urls += ['http://www.learntarot.com/c' + card + '.htm' for card in minorOrder]
# Swords
urls += ['http://www.learntarot.com/s' + card + '.htm' for card in minorOrder]
# Pentacles
urls += ['http://www.learntarot.com/p' + card + '.htm' for card in minorOrder]
# Extracts html
htmls = [requests.get(url).content for url in urls]
```

```python
# Clears all tags
texts = [html2text.html2text(html) for html in htmls]
```

```python
# Extract card titles
c = 77
card = re.search(r'\#\s([\w\s]*)\n\n', texts[c]).groups()[0]
print(card)

# Extracts card associations
association = re.findall(r'\*\s\*\*([\w\s\.]*)\*\*', texts[c])
print(association)
#opposing = re.findall(r'\[([\w\s]+)\]', re.findall(r'\(howcard\.htm\#howopposite\).*REINFORCING', texts[6],re.DOTALL)[0],re.DOTALL)
#print(opposing)
#reinforcing = re.findall(r'\[([\w\s]+)\]', re.findall(r'\(howcard\.htm\#howreinforce\).*\[DESCR', texts[6],re.DOTALL)[0],re.DOTALL)
#print(reinforcing)
```

```python
#Creates the dictionary of cards and associations

cards_and_associations = {re.search(r'\#\s([\w\s]*)\n\n', text).groups()[0] : 
                          {'Associations' : re.findall(r'\*\s\*\*([\w\s\.]*)\*\*', text)} for text in texts}

    
```

```python
# Fixes the difference in Knight's text
def fixKnight(k):
    assocs = []
    for s in k:
        assocs.extend(s.replace('..........',' ').split(' '))
    return assocs

cards_and_associations['KNIGHT OF CUPS']['Associations']   = fixKnight(cards_and_associations['KNIGHT OF CUPS']['Associations'])
cards_and_associations['KNIGHT OF SWORDS']['Associations'] = fixKnight(cards_and_associations['KNIGHT OF SWORDS']['Associations'])
cards_and_associations['KNIGHT OF PENTACLES']['Associations'] = fixKnight(cards_and_associations['KNIGHT OF PENTACLES']['Associations'])
cards_and_associations['KNIGHT OF WANDS']['Associations'] = fixKnight(cards_and_associations['KNIGHT OF WANDS']['Associations'])
```
